# Heatmaps Demo

### Load eICU CRD (Demo)

```python
import sys; sys.path.append('..')
from artemis.artemis import Artemis
from artemis.events import *
from artemis.thresholds import *

artemis = Artemis(
    '../../physionet.org/files/eicu-crd-demo/2.0.1',
    filter=lambda patient: patient.hospitalid == 59).load()
```

### Define Parameter and Thresholds

```python
parameter = 'heartrate'
params_fixed = {
    'high_threshold': range(80, 160, 10),
    'low_threshold': range(60, 120, 10)}
```

### Alarm Heatmap for a Single Patient

```python
ax = artemis.patients[0].parameters[parameter].heatmap(fixed, params_fixed, every)
ax.get_figure().savefig('../generated/grid_comparison_single_patient.pdf')
```

### Alarm Heatmap for All Patients

```python
ax = artemis.heatmap(parameter, fixed, params_fixed, every)
ax.get_figure().savefig('../generated/grid_comparison_all_patients.pdf')
```
